```python
MY_HF_TOKEN = input('ÌÜ†ÌÅ∞ÏùÑ ÏûÖÎ†•Ìï¥Ï£ºÏÑ∏Ïöî')
```

```python
import transformers
import torch
```

```python
model_id = 'meta-llama/Llama-3.2-1B-Instruct'
```

```python
pipeline = transformers.pipeline(
    'text-generation',
    model = model_id,
    model_kwargs = {'torch_dtype': torch.bfloat16},  # Î™®Îç∏ Ï¥àÍ∏∞Ìôî Ïãú ÏÑ§Ï†ï Í∞í (GPUÏóêÏÑú ÌÅ∞ Î™®Îç∏ ÏÇ¨Ïö© Ïãú Ï£ºÎ°ú ÏÇ¨Ïö©ÎêòÎäî Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ: bfloat16)
    device_map = 'auto',  # GPUÏôÄ CPUÎ•º ÏûêÎèôÏ†ÅÏúºÎ°ú Ìö®Ïú®Ï†ÅÏúºÎ°ú Î∞∞Ïπò
    token = MY_HF_TOKEN
)
```

```python
messages = [
    {'role': 'system',
     'content': 'You are a helpful AI assistant who answers question in a friendly and informal manner.'},
    {'role': 'user',
     'content': 'Tell us about Transformers self-attention.'}
]
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
chat_temp = tokenizer.apply_chat_template(messages, add_generation_prompt=True)
```

```python
tokenizer.decode(chat_temp)
```

```python
terminators = [
    pipeline.tokenizer.eos_token_id,
    pipeline.tokenizer.convert_tokens_to_ids('<|eot_id|>')
]
```

```python
outputs = pipeline(
    messages,
    max_new_tokens = 512,
    eos_token_id = terminators,
    do_sample = True,
    temperature = 0.6,  # ÏÉòÌîåÎßÅ Í≥ºÏ†ïÏóêÏÑú ÎûúÎç§ÏÑ±ÏùÑ Ï°∞Ï†à (Í∞íÏù¥ ÎÜíÏùÑÏàòÎ°ù Ï∞ΩÏùòÏ†ÅÏù¥Í≥† Îã§ÏñëÌïú ÏùëÎãµÏùÑ ÏÉùÏÑ±)
    top_p = 0.9,  # ÎàÑÏ†Å ÌôïÎ•†Ïù¥ 0.9 Ïù¥ÏÉÅÏù∏ Í≤ÉÎßå ÌôúÏö©ÌïòÍ≤†Îã§
)
```

```python
print(outputs[0]['generated_text'][-1]['content'])
```
